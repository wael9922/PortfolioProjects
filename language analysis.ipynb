```python
df.set_index('Respondent',inplace=True) #setting the Respondent column as index for the data frame
```

```python
grp = df.groupby(['Country']) #create data frame group by country object
```

```python
grp 
```

```python
#how to get any of the group you created (each country is considered a group)
grp.get_group(('India',)) #return the survey answers for each participant from India
```

```python
#return  a series of the number of survey participant for each country
particepant_for_group = grp['Respondent'].count() 
#return  a series of the number of survey participant who know Python for each country
py = grp['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').sum())
#return  a series of the number of survey participant who know Java for each country
java =grp['LanguageWorkedWith'].apply(lambda x: x.str.contains('Java').sum())
#return  a series of the number of survey participant who know C++ for each country
C =grp['LanguageWorkedWith'].apply(lambda x: x.str.contains('C++').sum())
#return  a series of the number of survey participant who know HTML for each country
html =grp['LanguageWorkedWith'].apply(lambda x: x.str.contains('HTML').sum())
```

```python
programming_languages_df = pd.concat([particepant_for_group,py,java,C,html],axis='columns') #combining the results into one data frame
programming_languages_df.columns = ['Number of Participants','Python','Java','C++','HTML'] #renaming the columns in the data frame
```
